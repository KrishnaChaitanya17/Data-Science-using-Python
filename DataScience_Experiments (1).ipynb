# 2.4.Write a python program to read a Blood_Pressure.csv file which is uploaded in the editch and perform the  Paired sampled t-test (blood pressure for an individual before and after some treatment) 

# H0:- means difference between two sample is 0 

# H1:- mean difference between two sample is not 0 

```
df=pd.read_csv("Blood_Pressure.csv")
df
```

```
from scipy.stats import ttest_rel
ttest,pval=ttest_rel(df['bp_before'],df['bp_after'])
print(pval)#alpha-5%
if pval<0.05:
  print("reject null hypothesis")
else:
  print("we fail to reject hypothesis")  
```

# 4.1.Generate the 10000 random numbers using normal distribution mean=0 and variance =1,(use norm.rvs function) and plot the normal distributions using seaborn

```
import numpy as np
import seaborn as sns
from scipy.stats import norm
#np.random.seed(42)
data_normal=norm.rvs(size=10000,loc=0,scale=1)#size,loc-mean,scale-variance

ax=sns.distplot(data_normal,bins=100,kde=True,color='skyblue',hist_kws={"linewidth":1,'alpha':1})
ax.set(xlabel='Normal',ylabel='Frequency')
```
